1. take inout of two numbers from the user
2. ask for the operation, 1 for addition, 2 for subtract, 3 for multiply, 4 for division
3. perform the operation and print the result

```python
first_number = int(input("Enter first number : "))
Second_number = int(input("Enter Second number : "))

operation_type=int(input("Enter your operation number : 1 for addition, 2 for subtract, 3 for multiply, 4 for division"))
```

```python
def addition(first_number,Second_number):
    add = first_number+Second_number
    return add

def subtraction(first_number,Second_number):
    Sub = first_number-Second_number
    return Sub
    
def multiplication(first_number,Second_number):
    mult=first_number*Second_number
    return mult

def Division(first_number,Second_number):
    div=first_number/Second_number
    return div
```

```python
if operation_type==1:
    print(addition(first_number,Second_number))

elif operation_type==2:
    print(subtraction(first_number,Second_number))

elif operation_type==3:
    print(multiplication(first_number,Second_number))

elif operation_type==4:
    print(Division(first_number,Second_number))

else:
    print("invalid input") 
 
```
