```python
# Exercise 1: Simple Tip Calculator

# Ask user for bill amount
bill_amount = float(input("Enter the bill amount: "))

# Ask user for tip percentage
tip_percent = float(input("Enter the tip percentage: "))

# Calculate tip amount
tip_amount = bill_amount * tip_percent / 100

# Calculate total bill
total_amount = bill_amount + tip_amount

# Display results formatted to 2 decimal places
print(f"Tip amount: ${tip_amount:.2f}")
print(f"Total amount: ${total_amount:.2f}")
```

```python
# Exercise 2: Temperature Converter

# Ask user for temperature in Celsius
celsius = float(input("Enter temperature in Celsius: "))

# Convert to Fahrenheit
fahrenheit = celsius * 9/5 + 32

# Display both temperatures formatted nicely
print(f"Temperature in Celsius: {celsius:.2f}°C")
print(f"Temperature in Fahrenheit: {fahrenheit:.2f}°F")
```
