# Write a function `count_long_lines(filename, limit)` that:
# 1. Opens a text file and reads all lines.
# 2. Uses a **for-loop** to count how many lines have length > limit.
# 3. Uses an **if-statement** inside the loop to check line length.
# 4. Returns the count.
# Then, write code that:
# - Prompts the user for a filename and a limit.
# - Calls the function.
# - Prints the result.

```python
def count_long_lines(filename, limit):
    """
    Counts how many lines in a file have length > limit.
    
    Args:
        filename (str): Name of the text file to read
        limit (int): The line length threshold
        
    Returns:
        int: Number of lines with length > limit
    """
    count = 0
    
    try:
        with open(filename, 'r') as file:
            for line in file:
                # Remove newline character for accurate length check
                line = line.strip()
                if len(line) > limit:
                    count += 1
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
        return 0
    except Exception as e:
        print(f"Error reading file: {e}")
        return 0
    
    return count


# Get user input
filename = input("Enter the filename: ")
limit = int(input("Enter the line length limit: "))

# Call the function
result = count_long_lines(filename, limit)

# Print the result
print(f"\nNumber of lines with length > {limit}: {result}")
```

# Create a function `filter_words(input_file, output_file, min_length)` that:
# 1. Opens input_file, reads all lines.
# 2. For each line, splits into words.
# 3. Uses a loop + if-statement to keep only words whose length >= min_length.
# 4. Writes these filtered words into output_file (one line per original line).
# Then call the function using any sample input/output filenames.

```python
def filter_words(input_file, output_file, min_length):
    try:
        with open(input_file, 'r') as infile:
            lines= infile.readlines()
        with open(output_file, 'w') as outfile:
            for line in lines:
                words = line.split()
                filtered_words = []
                for word in words:
                    if len(word) >= min_length:
                        filtered_words.append(word)
                # Write the filtered words as a line
                outfile.write(''.join(filtered_words) + '\n')
        print(f"Filtered words written to {output_file}")
    except FileNotFoundError:
        print(f"Error: Input file '{input_file}' not found.")
    except Exception as e:
        print(f"Error: {e}")

# Create a sample input file for demonstration
with open('sample_input.txt', 'w') as f:
    f.write("This is a sample line with short and long words.\n")
    f.write("Another line here with various word lengths.\n")
    f.write("Python programming is fun and educational.\n")

# Call the function with min_length=4
filter_words('sample_input.txt', 'filtered_output.txt', 4)

# Read and display the output file
print("\nContents of filtered_output.txt:")
with open('filtered_output.txt', 'r') as f:
    print(f.read())
```

```python
def filter_words(input_file, output_file, min_length):
    # Open input file for reading
    with open(input_file, "r") as infile:
        lines = infile.readlines()

    # Open output file for writing
    with open(output_file, "w") as outfile:
        for line in lines:
            words = line.split()   # split line into words
            
            filtered_words = []
            
            for word in words:
                if len(word) >= min_length:
                    filtered_words.append(word)
            
            # join filtered words and write to file
            outfile.write(" ".join(filtered_words) + "\n")

# Example usage
output_file = "output.txt"

filter_words("input_file.txt", output_file, 6)
print(f"Filtered words written to {output_file}")

with open("output.txt", "r") as f:
    print(f.read())
```

# Write code that:
# 1. Opens a file "temperatures.txt" which contains one temperature per line.
# 2. Reads all lines using readlines().
# 3. Converts them to floats.
# 4. Uses a function `average_temp(data)` to compute the average.
# 5. Prints:
# - The average temperature
# - "High" if avg > 25
# - "Moderate" if 15 <= avg <= 25
# - "Low" if avg < 15

```python
def average_temp(data):
    """
    Computes the average of a list of numbers.
    
    Args:
        data (list): List of float numbers
        
    Returns:
        float: The average of the numbers
    """
    if not data:
        return 0.0
    return sum(data) / len(data)

# Open the file and read all lines
with open("temperatures.txt", "r") as file:
    lines = file.readlines()

# Convert lines to floats
temperatures = []
for line in lines:
    try:
        temp = float(line.strip())
        temperatures.append(temp)
    except ValueError:
        print(f"Warning: Skipping invalid temperature: {line.strip()}")

# Compute the average
avg = average_temp(temperatures)

# Print the average
print(f"The average temperature is: {avg:.2f}")

# Classify the temperature
if avg > 25:
    print("High")
elif 15 <= avg <= 25:
    print("Moderate")
else:
    print("Low")
```

# Exercise 5:

# Write a program that:
# 1. Opens a file "logs.txt" and reads each line.
# 2. Each line is formatted like: "LEVEL: message"
# 3. Use a function `count_level(filename, level)` that:
# - Opens the file
# - Counts how many lines start with the given level (e.g., "ERROR")
# - Uses loops + if to check matching lines
# 4. Print how many ERROR, WARNING, and INFO lines are in the file.

```python
def count_level(filename, level):
    """
    Counts how many lines in the file start with the given level.
    
    Args:
        filename (str): Name of the log file
        level (str): The level to count (e.g., "ERROR", "WARNING", "INFO")
        
    Returns:
        int: Number of lines starting with the level
    """
    count = 0
    try:
        with open(filename, 'r') as file:
            for line in file:
                if line.startswith(level + ":"):
                    count += 1
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
        return 0
    except Exception as e:
        print(f"Error reading file: {e}")
        return 0
    return count

# Count each level
error_count = count_level("logs.txt", "ERROR")
warning_count = count_level("logs.txt", "WARNING")
info_count = count_level("logs.txt", "INFO")

# Print the results
print(f"Number of ERROR lines: {error_count}")
print(f"Number of WARNING lines: {warning_count}")
print(f"Number of INFO lines: {info_count}")
```
