# Exercise 3: 
# Create a list of 3 coordinate tuples (x, y)
# Print the first and last coordinates
# Create and print a new list with x and y swapped in each tuple

```python
# Create a list of 3 coordinate tuples (x, y)
coordinates = [(1, 2), (3, 4), (5, 6)]

# Print the first and last coordinates
print("First coordinate:", coordinates[0])
print("Last coordinate:", coordinates[-1])

for index, (x, y) in enumerate(coordinates):
    print(f"Coordinate {index}: x={x}, y={y}")

for x,y in coordinates:
    print(f"x={x}, y={y}")

# Create and print a new list with x and y swapped in each tuple
swapped_coordinates = [(y,x) for x,y in coordinates]
print("Swapped coordinates:", swapped_coordinates)
```

# Exercise 4:
# Create two sets: tech_subscribers and news_subscribers with some overlapping emails
# Print the emails that are in both sets
# Print all unique emails from both sets
# Remove one email from both sets and print the updated sets

```python
# Create two sets with some overlapping emails
tech_subscribers = {"alice@example.com", "bob@example.com", "charlie@example.com"}
news_subscribers = {"bob@example.com", "charlie@example.com", "diana@example.com"}

# Print the emails that are in both sets (intersection)
common_emails = tech_subscribers & news_subscribers
print("Emails in both sets:", common_emails)

# Print all unique emails from both sets (union)
all_unique_emails = tech_subscribers | news_subscribers
print("All unique emails:", all_unique_emails)

# Remove one email from both sets and print the updated sets
email_to_remove = "bob@example.com"
tech_subscribers.discard(email_to_remove)
news_subscribers.discard(email_to_remove)
print("Updated tech_subscribers:", tech_subscribers)
print("Updated news_subscribers:", news_subscribers)

# Remove all overlapping emails from both sets
tech_subscribers -= common_emails
news_subscribers -= common_emails

# Print the updated sets
print("Updated tech_subscribers1:", tech_subscribers)
print("Updated news_subscribers1:", news_subscribers)
```

# Exercise 5: 
# Create a dictionary "library" with 3 books; each value is a dictionary with "title", "author", "available"
# Create a list "borrowing_history" with 2 tuples (book_id, user_name)
# Print all titles of books that are available
# Print all user names that appear in borrowing_history

```python
# Create a dictionary "library" with 3 books; each value is a dictionary with "title", "author", "available"
library = {
    "book1": {"title": "Python Programming", "author": "John Doe", "available": True},
    "book2": {"title": "Data Engineering", "author": "Jane Smith", "available": False},
    "book3": {"title": "Machine Learning", "author": "Bob Johnson", "available": True}
}

# Create a list "borrowing_history" with 2 tuples (book_id, user_name)
borrowing_history = [("book1", "Alice"), ("book2", "Bob")]

# Print all titles of books that are available
print("Available book titles:")
for book_id, book_info in library.items():
    if book_info["available"]:
        print(book_info["title"])

# Print all user names that appear in borrowing_history
print("\nUser names in borrowing history:")
users = set(user for _, user in borrowing_history)
for user in users:
    print(user)
```
